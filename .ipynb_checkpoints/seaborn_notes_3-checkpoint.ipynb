```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
tips = sns.load_dataset('tips')
```

```python
iris.head()
```

```python
tips.head()
```

<a id='pairplot'></a>
### Pairplot
[Back to top](#top)

```python
sns.pairplot(tips,hue='time')
plt.show()
```

<a id='catplot'></a>
### Catplot
[Back to top](#top)

```python
sns.catplot(x='sex',y='tip',col='day',col_wrap=2,data=tips,kind='swarm')
plt.show()
```

```python
sns.catplot(x='sex',y='tip',col='day',col_wrap=2,data=tips,kind='box')
plt.show()
```

<a id='facetgrid'></a>
### FacetGrid
[Back to top](#top)

```python
fgrid = sns.FacetGrid(tips, col = "day",hue='sex',col_wrap=2)
fgrid.map(sns.scatterplot,'total_bill','tip')
plt.show()
```

```python
fgrid = sns.FacetGrid(tips, col = "day",col_wrap=2)
fgrid.map(sns.distplot,'tip')
plt.show()
```

<a id='pairgrid'></a>
### PairGrid
[Back to top](#top)

```python
pgrid = sns.PairGrid(tips)
pgrid.map_diag(sns.kdeplot,lw=3,color='red')
pgrid.map_upper(sns.scatterplot,color='orange')
pgrid.map_lower(sns.kdeplot,cmap='inferno')
plt.show()
```

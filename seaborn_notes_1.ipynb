### Styling
[Back to top](#top)

```python
# matplotlib to seaborn

x = np.arange(1,21)
y1 = x*2
y2 = x*3

print(x)
print(y1)
print(y2)
```

```python
plt.figure(figsize=(8,5))
plt.plot(x,y1)
plt.plot(x,y2)
plt.show()
```

```python
sns.set()
plt.figure(figsize=(8,5))
plt.plot(x,y1)
plt.plot(x,y2)
plt.show()
```

```python
# set_style()
# darkgrid, whitegrid, dark, white, ticks

# set_context()
# paper, notebook, talk, poster

plt.plot(x,y1)
plt.plot(x,y2)
sns.set_style("darkgrid") 
sns.set_context("talk")
sns.despine()
plt.show()
```

```python
sns.axes_style()
```

```python
plt.plot(x,y1)
plt.plot(x,y2)
sns.set_style({'grid.color': '.9'})
sns.set_context(None)
plt.show()
```

```python
sns.palplot(sns.color_palette())
```

```python
sns.palplot(sns.color_palette("BrBG", 7))
```

```python
plt.figure(figsize=(12,8))
sns.relplot(x="age",y="fare",data=titanic,hue='sex',palette="inferno")
plt.show()
```

```python
sns.relplot(x="age",y="fare",data=titanic,col="class",hue='sex',col_wrap=2,palette="inferno")
plt.show()
```

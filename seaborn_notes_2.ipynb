```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic = sns.load_dataset('titanic')
iris = sns.load_dataset('iris')
tips = sns.load_dataset('tips')
```

<a id='barplot'></a>
### Barplot
[Back to top](#top)

```python
sns.barplot(x='pclass',y='age',data=titanic,hue='sex',ci=None,estimator=np.mean)
plt.show()
```

```python
sns.barplot(x='petal_length',y='species',data=iris,ci=None,estimator=sum,orient='h')
plt.show()
```

<a id='pointplot'></a>
### Pointplot
[Back to top](#top)

```python
sns.pointplot(x='pclass',y='age',data=titanic,hue='sex',estimator=np.mean)
plt.show()
```

<a id='countplot'></a>
### Countplot
[Back to top](#top)

```python
sns.countplot(x='sex',data=titanic)
plt.show()
```

<a id='histogram'></a>
### Histogram
[Back to top](#top)

```python
sns.distplot(titanic['age'].dropna(),bins=8,kde=False,color='b')
plt.show()
```

<a id='kde'></a>
### KDE
[Back to top](#top)

```python
# kde using displot

sns.distplot(iris[iris['species']=='setosa']['petal_length'],hist=False,bins=5,color='b',label='setosa petal length')
sns.distplot(iris[iris['species']=='setosa']['petal_width'],hist=False,bins=5,color='r',label='setosa petal width')
```

```python
setosa_petal_length = iris[iris['species']=='setosa']['petal_length']
setosa_petal_width = iris[iris['species']=='setosa']['petal_width']
versicolor_petal_length = iris[iris['species']=='versicolor']['petal_length']
versicolor_petal_width = iris[iris['species']=='versicolor']['petal_width']
virginica_petal_length = iris[iris['species']=='virginica']['petal_length']
virginica_petal_width = iris[iris['species']=='virginica']['petal_width']

sns.kdeplot(setosa_petal_length,color='b',shade=True,label='setosa petal length')
sns.kdeplot(setosa_petal_width,color='r',shade=True,label='setosa petal width')
plt.show()
```

```python
sns.kdeplot(setosa_petal_length,setosa_petal_width,label='setosa petal',cmap='Blues')
sns.kdeplot(versicolor_petal_length,versicolor_petal_width,label='versicolor petal',cmap='Oranges')
sns.kdeplot(virginica_petal_length,virginica_petal_width,label='virginica petal',cmap='Greens')
plt.legend()
plt.title('Iris species')
plt.show()
```

<a id='boxplot'></a>
### Box Plot
[Back to top](#top)

```python
sns.boxplot(x='day',y='tip',hue='sex',data=tips,hue_order=['Male','Female'],width=0.6)
plt.show()
```

<a id='violinplot'></a>
### Violin Plot
[Back to top](#top)

```python
sns.violinplot(x='day',y='tip',hue='sex',data=tips,palette='inferno')
plt.show()
```

<a id='scatterplot'></a>
### Scatter Plot
[Back to top](#top)

```python
sns.relplot(x="total_bill", y="tip",hue='sex',data=tips)
plt.show()
```

```python
sns.regplot(x="total_bill", y="tip",data=tips)
plt.show()
```

```python
# regplot with FacetGrid

sns.lmplot(x="total_bill", y="tip",row='time',col='sex',hue='sex',data=tips)
plt.show()
```

<a id='jointplot'></a>
### Joint Plot
[Back to top](#top)

```python
sns.jointplot(x="total_bill", y="tip",data=tips)
plt.show()
```

```python
sns.jointplot(x="total_bill", y="tip",kind='reg',data=tips,marginal_kws=dict(bins=15, rug=True))
plt.show()
```

```python
sns.jointplot(x="total_bill", y="tip",kind='hex',data=tips,color='g')
plt.show()
```

```python
sns.jointplot(x="total_bill", y="tip",kind='kde',data=tips)
plt.show()
```
